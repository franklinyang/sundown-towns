```python
data_dir = os.path.join(os.path.abspath(os.path.dirname('.')), '../data')
```

```python
police_killings_data = pd.read_csv(os.path.join(data_dir, 'outputs/police_killings_county.csv'))
police_killings_data = police_killings_data.rename(columns={police_killings_data.columns[0]: 'id'})
police_killings_data = police_killings_data.drop(columns=police_killings_data.columns[1])
sundown_towns_county = pd.read_csv(os.path.join(data_dir, 'outputs/county_sundown_counts.csv'))
```

```python
shootings_in_sundown_towns = police_killings_data.merge(sundown_towns_county, on='county', how='left')
shootings_in_sundown_towns['num_sundown_towns'] = shootings_in_sundown_towns['num_sundown_towns'].fillna(0)
```
